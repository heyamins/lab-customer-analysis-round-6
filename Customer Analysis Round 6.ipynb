```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


import pickle as pkl

import warnings
warnings.filterwarnings('ignore') 
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('./files_for_lab/csv_files/marketing_customer_analysis.csv')
df
```

```python
# x y split

X = df.drop(["Total Claim Amount"], axis = 1)
y = df["Total Claim Amount"]
```

```python
X.select_dtypes(np.number)
```

```python
scaler = StandardScaler()
```

```python
#Normalize (numerical).
X_num = X.select_dtypes(np.number)

scaler = StandardScaler()
scaler.fit(X_num)
X_num_scaled = scaler.transform(X_num)
X_num_scaled = pd.DataFrame(X_num_scaled, columns = X_num.columns)
```

#### Format Transformation: 
The scaler.transform(X_num) method returns a NumPy array, not a DataFrame. While NumPy arrays are efficient for calculations, they don't have the convenient data structure and labeling that a DataFrame provides.

#### Preserving Column Names:
The columns = X_num.columns part ensures that the column names from the original DataFrame (X_num) are used in the new DataFrame (X_num_scaled). This makes it easier to understand which column corresponds to which feature.

Before make one hot encoding, I need to remove not necessary columns

```python
X = X.drop(['Customer'], axis = 1)
```

```python
X_cat = X.select_dtypes(include=['object'])
X_cat = pd.get_dummies(X_cat)
```

```python
X_cat
```

Concatenate DataFrames

```python
X_new = pd.concat([X_num_scaled, X_cat], axis = 1 )
```

```python
model = LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size = 0.2, random_state=42)
```

```python
model.fit(X_train, y_train)
```

```python
#I splited the X_new data and y data y is target. test_size is 0.2
#now 'model' is LinearRegression model object that X_train, y_train is trained.
```

```python
#for R2, MSE, RMSE, MAE I need these libraries.

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
#model has predict function and I can put X_test values to test those.
y_prediction = model.predict(X_test)

r2 = r2_score(y_test, y_prediction)
mse = mean_squared_error(y_test, y_prediction)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_prediction)
```

```python
print(r2, mse, rmse, mae)
```
